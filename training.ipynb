```python
from sklearn.model_selection import train_test_split
target_col = ['rfh', 'rfh_avg', 'r1h', 'r1h_avg', 'r3h', 'r3h_avg', 'rfq', 'r1q', 'r3q']
X=df.drop(target_col, axis =1) #feature
#axis mean the tabular form we are dropping; 0 for row and 1 for col
y=df[target_col] #target
```
